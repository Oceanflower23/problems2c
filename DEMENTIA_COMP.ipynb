DEMENTIA DATA DISTRIBUTION AND CONFIDENCE INTERVALS


```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
from scipy import stats

folder_path = "/content/DEMENTIA"
columns = ["Auditory.mean", "Gustatory.mean", "Haptic.mean", 
           "Interoceptive.mean", "Olfactory.mean", "Visual.mean", 
           "Foot_leg.mean", "Hand_arm.mean", "Head.mean", 
           "Mouth.mean", "Torso.mean"]
dfs = {}
# Iterate
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'): 
        df = pd.read_csv(os.path.join(folder_path, filename))

# REMOVE NaN
        df = df.dropna(subset=columns)
        if len(df) > 0:
            dfs[filename] = df
combined_df = pd.concat(dfs.values(), axis=0, ignore_index=True)

# Histogram
for column in columns:
    plt.figure(figsize=(10, 6))
    sns.distplot(combined_df[column], kde=False, rug=True, bins=30,
                 hist_kws=dict(edgecolor="k", linewidth=2))

# Confidence intervals
    ci = stats.norm.interval(0.95, combined_df[column].mean(), combined_df[column].std())
    plt.axvline(ci[0], color='red', linestyle='dashed', linewidth=2)
    plt.axvline(ci[1], color='red', linestyle='dashed', linewidth=2)

    plt.title(f'Histogram of {column} with 95% Confidence Intervals')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

T TEST 

```python
from scipy import stats
dir_dementia = "/content/DEMENTIA"
dir_control = "/content/CONTROL"

columns = ["Auditory.mean", "Gustatory.mean", "Haptic.mean", 
           "Interoceptive.mean", "Olfactory.mean", "Visual.mean", 
           "Foot_leg.mean", "Hand_arm.mean", "Head.mean", 
           "Mouth.mean", "Torso.mean"]

#CREATE DF
df_dementia = pd.DataFrame(columns=columns)
df_control = pd.DataFrame(columns=columns)

# LOAD THE FILES FROM DEMENTIA DIRECTORY
for file in os.listdir(dir_dementia):
    df_temp = pd.read_csv(os.path.join(dir_dementia, file))
    df_dementia = pd.concat([df_dementia, df_temp])

# SAME FOR CONTROL DIRECTORY
for file in os.listdir(dir_control):
    df_temp = pd.read_csv(os.path.join(dir_control, file))
    df_control = pd.concat([df_control, df_temp])

# WE NOW CALCULATE THE T-TESTS
for column in columns:
    t_stat, p_val = stats.ttest_ind(df_dementia[column].dropna(), df_control[column].dropna())
    print(f"For {column}: t-statistic = {t_stat:.3f}, p-value = {p_val:.3f}")
```

MULTIPLE REGRESSION - DEPENDENT:OLFACTORY MEAN
INDEPENDENT VARIABLES:AUDITORY, GUSTATORY AND INTEROCEPTIVE MEANS

```python
import pandas as pd
import statsmodels.api as sm

data = pd.read_csv("/content/combined_dementia_data.csv")

# our independent variables
independent_vars = ["Auditory.mean", "Gustatory.mean", "Interoceptive.mean"]

# constant term to dependant variable
data["intercept"] = 1

# design matrix
X = data[["intercept"] + independent_vars]

# dependent vatiable on the intercept
y = data["Olfactory.mean"]

#The multiple linear regression model fit
model = sm.OLS(y, X)
results = model.fit()

print(results.summary())
```
